```python
import numpy as np
from sklearn.linear_model import LinearRegression

# 1. Prepare Data (X = Hours studied, y = Exam Score)
# We use .reshape(-1, 1) because Scikit-Learn expects a 2D grid for X
X = np.array([1, 2, 3, 4, 5]).reshape(-1, 1) 
y = np.array([45, 55, 65, 75, 85])

# 2. Initialize the "Brain" (The Model)
model = LinearRegression()

# 3. Train the Brain (This is called "Fitting" the model)
# The computer looks at X and y and calculates the straight line
model.fit(X, y)

# 4. Make a Prediction
# Let's guess the score for a student who studies 6 hours
hours_new = np.array([[6]])
prediction = model.predict(hours_new)

print(f"Prediction for 6 hours of study: {prediction[0]}%")
```

```python
import pandas as pd
import joblib
import category_encoders as ce
from sklearn.pipeline import Pipeline
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import MinMaxScaler
from sklearn.model_selection import train_test_split

# 1. Load Dataset
data = pd.read_csv("House_Rent_Dataset.csv")

# 2. Separate Features and Target
X = data.drop(columns=["Rent", "Posted On", "Floor", "Area Locality"])
y = data["Rent"]

# 3. Train-Test Split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=42)

# 4. Build the Pipeline (This glues everything together!)
pipeline = Pipeline([
    ("encoder", ce.LeaveOneOutEncoder()),
    ("scaler", MinMaxScaler()),
    ("model", LinearRegression())
])

# 5. Train the entire Pipeline at once
pipeline.fit(X_train, y_train)

# 6. Save the entire Pipeline into a single .pkl file
joblib.dump(pipeline, "house_rent_pipeline.pkl")
print("Pipeline saved successfully as house_rent_pipeline.pkl!")
```

```python
import pandas as pd
data = pd.read_csv("House_Rent_Dataset.csv")
y=data['Rent']
x=data.drop(['Rent','Posted On','Floor','Area Locality'], axis=1)
import category_encoders as ce
encoder = ce.LeaveOneOutEncoder(return_df=True)
x = encoder.fit_transform(x, y)
import pandas
x = pd.DataFrame(x, columns=x.columns)
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
```

```python
import tensorflow as tf
from tensorflow.keras.models import Sequential  
from tensorflow.keras.layers import Dense
model = Sequential()
model.add(Dense(32, input_shape=(x_train.shape[1],), activation='relu'))
model.add(Dense(1, activation='relu'))  # Output layer for regression
model.compile(optimizer='adam', loss='mean_squared_error')
model.fit(x_train, y_train, epochs=100, batch_size=32, validation_data=(x_test, y_test))
```

```python
import numpy as np

# 1. Test dataset (x_test) ke sabhi gharon ke liye Rent predict karna
predictions = model.predict(x_test)

# 2. Pehle 5 gharon ka Predicted Rent aur Asli (Actual) Rent aamne-saamne dekhna
print("--- first 5 house's Rent Comparison ---")
for i in range(5):
    print(f"House {i+1} -> Predicted Rent: {predictions[i][0]:.2f} | Actual Rent: {y_test.values[i]}")
```
