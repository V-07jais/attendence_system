```python
import numpy as np
import matplotlib.pyplot as plt

# 1. Ek 16x16 size ki safed (white) photo banana jisme 3 channels (RGB) hain
# np.full se computer ek matrix banayega jisme har jagah 255 (yani white color) bhara hoga
image = np.full((16, 16, 3), 255, dtype="uint8")

# 2. Row 4 se 10, aur Column 4 se 8 ke beech ke pixels ko LAL (Red) karna
# [255, 0, 0] matlab Red full, Green zero, Blue zero
image[4:10, 4:8] = [255, 0, 0]

# 3. Row 4 se 10, aur Column 8 se 11 ke beech ke pixels ko PEELA (Yellow) karna
# [255, 255, 0] matlab Red aur Green full, jisse Yellow color banta hai
image[4:10, 8:11] = [255, 255, 0]

# 4. Photo ko screen par plot karke dekhna
# cmap='jet' sir ne lagaya hai, par simple imshow bhi chalega
plt.imshow(image, cmap='jet')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# 1. Flag ka shape define karna (Row=12, Column=18 ka ek ratio le rahe hain)
# 3 channels (RGB) ke sath white canvas (255) banana
flag = np.full((12, 18, 3), 255, dtype="uint8")

# 2. Top Band: Row 0 se 4 tak ko Saffron (Keshari) rang dena
# RGB value for Saffron roughly: [255, 153, 51]
flag[0:4, :] = [255, 153, 51]

# 3. Middle Band: Row 4 se 8 tak pehle se hi pure white [255, 255, 255] hai.

# 4. Bottom Band: Row 8 se 12 tak ko Green (Hara) rang dena
# RGB value for Green roughly: [19, 136, 8]
flag[8:12, :] = [19, 136, 8]

# 5. Tirange ko screen par bina kisi grid lines ke saaf-saaf dikhana
plt.imshow(flag)
plt.axis('on') # Isse graph ki numbering chhup jayegi aur flag ekdum real lagega
plt.show()
```

```python
import cv2
import numpy as np

# Create a sample image (500x500 with gradient colors)
image = np.zeros((500, 500, 3), dtype=np.uint8)
for i in range(500):
    for j in range(500):
        image[i, j] = [i % 256, j % 256, (i+j) % 256]

# Save as demo.jpg
cv2.imwrite('demo.jpg', image)
print("demo.jpg created successfully!")
```

```python
image=cv2.imread('demo.jpg',0)
image=cv2.GaussianBlur(image,(5,5),0)
threshold_image,th=cv2.threshold(image,157,255,cv2.THRESH_TOZERO)
cv2.imshow('image',th)
cv2.waitKey(1000)
cv2.destroyAllWindows()
```

```python
edge_detected_image=cv2.Canny(image,100,200)
cv2.imshow('image',edge_detected_image)
cv2.waitKey(10000)
cv2.destroyAllWindows()
```

```python
import cv2
import numpy as np

# 1. Apne folder ki maujooda photo ka naam likho (e.g., lio2.jpg)
image = cv2.imread('demo.jpg')

model = cv2.CascadeClassifier(cv2.data.haarcascades + 'haarcascade_frontalface_default.xml')
faces = model.detectMultiScale(image)
print(faces)

for face in faces:
    x, y, w, h = face
    # Blue rang ka dabba banane ke liye (255, 0, 0)
    cv2.rectangle(image, (x, y), (x + w, y + h), (255, 0, 0), 3)

# 2. Sahi photo par rectangle draw hone ke baad output.jpg save hogi
cv2.imwrite('output.jpg', image)
```
