```python
import cv2
from ultralytics import YOLO

# 1. YOLOv11 ka pretrained model load kiya
model = YOLO('yolo11n.pt')

# 2. Apni parking wali photo ka naam diya
img_path = 'parking.jpg'

# 3. Model ko bola ki photo mein sirf Car (2) aur Motorcycle (3) dhoondho
results = model(img_path, classes=[2, 3], conf=0.3)

# 4. YOLO ne jo dabbe banaye, unhe image par plot (draw) kiya
annotated_img = results[0].plot()

# 5. Backend se count nikalne ka jugaad:
# Bounding boxes ki details nikalte hain
boxes = results[0].boxes
car_count = 0
bike_count = 0

if boxes is not None:
    for box in boxes:
        # Har ek object ki class id check karte hain (2 = Car, 3 = Bike)
        cls_id = int(box.cls[0])
        if cls_id == 2:
            car_count += 1
        elif cls_id == 3:
            bike_count += 1

# 6. Screen par count print karo
print("\n" + "="*30)
print(f"📊 PARKING STATS:")
print(f"🚗 Total Cars Detected: {car_count}")
print(f"🏍️ Total Bikes Detected: {bike_count}")
print("="*30 + "\n")

# 7. Final output image ko save aur show karo
cv2.imwrite('parking_analysis.jpg', annotated_img)
cv2.imshow('Smart Parking System', annotated_img)
cv2.waitKey(7000) # 7 second tak photo dikhegi
cv2.destroyAllWindows()
```

```python
import cv2
from ultralytics import YOLO

model = YOLO('yolo11n.pt')
img_path = 'parking.jpg'

# Classes badal di: 1=Bicycle/Scooty filter, 2=Car, 3=Motorcycle. Conf thoda kam kiya.
results = model(img_path, classes=[1, 2, 3], conf=0.2)
annotated_img = results[0].plot()

boxes = results[0].boxes
car_count = 0
bike_count = 0

if boxes is not None:
    for box in boxes:
        cls_id = int(box.cls[0])
        if cls_id == 2:
            car_count += 1
        elif cls_id in [1, 3]:  # Agar 1 (bicycle) ya 3 (motorcycle) dono mein se kuch bhi miley
            bike_count += 1

print("\n" + "="*30)
print(f"📊 UPDATED PARKING STATS:")
print(f"🚗 Total Cars Detected: {car_count}")
print(f"🏍️ Total Bikes/Scooters Detected: {bike_count}")
print("="*30 + "\n")

cv2.imwrite('parking_analysis.jpg', annotated_img)
cv2.imshow('Smart Parking System', annotated_img)
cv2.waitKey(7000)
cv2.destroyAllWindows()
```
